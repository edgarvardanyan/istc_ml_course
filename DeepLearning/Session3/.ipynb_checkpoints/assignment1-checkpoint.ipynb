# Text Classification: sentiment analysis.

### Data

The dataset we will use of this problem is known as the Stanford Sentiment Treebank. Movie reviews with one sentence per review:
* `sst.bin.*` with binary labels: negative or positive
* `sst.fine.*` with fine-grained labels (very negative, negative, neutral, positive, very positive),

```python
import keras
import numpy as np
from matplotlib import pyplot as plt
```

```python
with open("data/sst.bin.train", "r") as f:
    train = [(int(x.split()[0]), x.split()[1:]) for x in f.readlines()]

with open("data/sst.bin.val", "r") as f:
    val = [(int(x.split()[0]), x.split()[1:]) for x in f.readlines()]
with open("data/sst.bin.test", "r") as f:
    test = [(x.split()[1:]) for x in f.readlines()]
print("Train:     ", len(train), "  Mean sentence length:", np.mean([len(x[1]) for x in train]).round(2))
print("Validation:", len(val))
print("Test:      ", len(test))
```

```python
with open("data/sst.bin.train", "r") as f:
    raw_sentences_train = [x[2:-1] for x in f.readlines()]
with open("data/sst.bin.val", "r") as f:
    raw_sentences_val = [x[2:-1] for x in f.readlines()]
with open("data/sst.bin.test", "r") as f:
    raw_sentences_test = [x[2:-1] for x in f.readlines()]
```

```python
from collections import Counter, defaultdict
import itertools
```

```python
# calculate word frequencies
word_frequencies = Counter(itertools.chain(
    *((word for word in sample[1]) for sample in train)
))
```

```python
print("Vocabulary size:", len(word_frequencies))

plt.plot([x[1] for x in word_frequencies.most_common()[::-1]])
plt.ylim(0, 20);
```

```python
# indexing words

PAD = 0 # padding
UNK = 1 # unknown word
vocab = dict()

word_index = 2
for w, c in word_frequencies.most_common():
    if c > 4:
        vocab[w] = word_index
        word_index += 1

len(vocab), word_index
```

```python
word_frequencies.most_common()
```

```python
# change words with indices

train_X = [[vocab.get(x, UNK) for x in sample[1]]
                    for sample in train]
train_y = [sample[0] for sample in train]

max_len = np.sort([len(x) for x in train_X])

val_X = [[vocab.get(x, UNK) for x in sample[1]]
                    for sample in val]
val_y = [sample[0] for sample in val]
```

```python
train_X = keras.preprocessing.sequence.pad_sequences(train_X, value=PAD)
train_y = np.array(train_y)

val_X = keras.preprocessing.sequence.pad_sequences(val_X, value=PAD)
val_y = np.array(val_y)

```

```python
train_X.shape, train_y.shape, val_X.shape, val_y.shape
```

```python
train_X[0]
```

```python
from keras import Sequential
from keras.layers import Embedding
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Dropout
```

```python
vocab
```

```python
# create the model
embedding_vecor_length = 32
model = Sequential()
model.add(Embedding(word_index, embedding_vecor_length))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
train_X[0]
```

```python
model.fit(train_X, train_y, validation_data=(val_X, val_y), epochs=10, batch_size=64)
```

## Naive Bayes

```python
#Naive Bayes
from sklearn.metrics import accuracy_score

def naive_bayes_on_text(test_X, train_X, train_y):
    test_y = np.zeros(test_X.shape[0], dtype = int)
    for i, review in enumerate(test_X):
        p_0 = 1
        p_1 = 1
        for word in review:
            if word>15: #throw away frequent words like 'a', 'the'
                p_0 *= (train_X[train_y == 0] == word).sum()/(train_y==0).sum()
                p_1 *= (train_X[train_y == 1] == word).sum()/(train_y==1).sum()
        if p_1>p_0:
            test_y[i] = 1
    return test_y

predicted_y = naive_bayes_on_text(val_X, train_X, train_y)

print("Accuracy score on validation set:", accuracy_score(predicted_y, val_y))
            
```

## Logistic Regression

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1', ngram_range=(1, 2), stop_words='english')
tfidf.fit(raw_sentences_train + raw_sentences_val)
train_features = tfidf.transform(raw_sentences_train).toarray()
val_features = tfidf.transform(raw_sentences_val).toarray()

a = LogisticRegression()
a.fit(train_features, train_y)

predicted_y = a.predict(val_features)

print("Accuracy score on validation set:", accuracy_score(predicted_y, val_y))
```

## Embedding + Convolution + Recurrent

```python
from keras.layers import Conv1D
from keras.layers import Convolution1D
from keras.layers import MaxPool1D
```

```python
embedding_vecor_length = 32
model = Sequential()
model.add(Embedding(word_index, embedding_vecor_length))
model.add(Conv1D(filters = 32, kernel_size = 3, padding = 'same', activation = 'relu'))
model.add(MaxPool1D(pool_size = 2))
model.add(LSTM(100))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(train_X, train_y, validation_data=(val_X, val_y), epochs=10, batch_size=64)
```

## Pretrained embedding + Convolution + LSTM

```python
from gensim.models import KeyedVectors
en_model = KeyedVectors.load_word2vec_format('wiki.simple.vec')
```

```python
def vectorize(word, vectorizer):
    try:
        vec = vectorizer[word]
        return vec
    except KeyError:
        return np.zeros(vectorizer.vector_size, dtype = np.float32)
```

```python
train_X = [[vectorize(x, en_model) for x in sample[1]]
                    for sample in train]
train_y = [sample[0] for sample in train]

val_X = [[vectorize(x, en_model) for x in sample[1]]
                    for sample in val]
val_y = [sample[0] for sample in val]
```

```python
pad = vectorize("zibil", en_model)
train_X = keras.preprocessing.sequence.pad_sequences(train_X, value=PAD, dtype = object, maxlen = 40)
train_y = np.array(train_y)

val_X = keras.preprocessing.sequence.pad_sequences(val_X, value=PAD, dtype = object, maxlen = 40)
val_y = np.array(val_y)
```

```python
model = Sequential()
model.add(Convolution1D(filters = 32, kernel_size = 3, padding = 'same', activation = 'relu', input_shape = (40,300)))
model.add(MaxPool1D(pool_size = 2))
model.add(LSTM(100, dropout = 0.3))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(train_X, train_y, validation_data=(val_X, val_y), epochs=15, batch_size=128)
```

## Pretrained embedding + LSTM

```python
from gensim.models import KeyedVectors
en_model = KeyedVectors.load_word2vec_format('wiki.simple.vec')
```

```python
def vectorize(word, vectorizer):
    try:
        vec = vectorizer[word]
        return vec
    except KeyError:
        return np.zeros(vectorizer.vector_size, dtype = np.float32)
```

```python
train_X = [[vectorize(x, en_model) for x in sample[1]]
                    for sample in train]
train_y = [sample[0] for sample in train]

val_X = [[vectorize(x, en_model) for x in sample[1]]
                    for sample in val]
val_y = [sample[0] for sample in val]

pad = vectorize("zibil", en_model)
train_X = keras.preprocessing.sequence.pad_sequences(train_X, value=PAD, dtype = object, maxlen = 40)
train_y = np.array(train_y)

val_X = keras.preprocessing.sequence.pad_sequences(val_X, value=PAD, dtype = object, maxlen = 40)
val_y = np.array(val_y)

```

```python
model = Sequential()
model.add(LSTM(100, input_shape = (40,300)))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

```python
model.fit(train_X, train_y, validation_data=(val_X, val_y), epochs=15, batch_size=64)
```
